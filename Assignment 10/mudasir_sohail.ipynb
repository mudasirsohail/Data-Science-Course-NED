# 62. Bar Chart (Students in Different Age Groups)

```python
age_groups = ['10-20', '21-30', '31-40', '41-50']
students = [50, 80, 40, 20]

plt.bar(age_groups, students, color='skyblue')
plt.title("Number of Students in Different Age Groups")
plt.xlabel("Age Groups")
plt.ylabel("Number of Students")
plt.show()
```

# 65. Pie Chart

```python
categories = ['Category A', 'Category B', 'Category C']
values = [20, 30, 50]

plt.pie(values, labels=categories, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue'])
plt.title("Category Distribution")
plt.show()
```

# 66. Two Subplots (Line and Scatter)

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4), sharex=True, sharey=True)

# Line plot
ax1.plot(x, y, color='blue')
ax1.set_title("Line Plot")

# Scatter plot
ax2.scatter(x, y, color='orange')
ax2.set_title("Scatter Plot")

plt.suptitle("Two Subplots Example")
plt.show()
```

# 67. Customized Line Plot

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, color='red', linewidth=2)
plt.title("Customized Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

# Seaborn Tasks

```python
# Create random dataset
data = pd.DataFrame({
    'x': np.arange(1, 11),
    'y': np.random.randint(10, 100, 10),
    'category': np.random.choice(['A', 'B', 'C'], 10)
})
```

# 68(a). Pie Chart (with Seaborn workaround)

```python
plt.figure(figsize=(5, 5))
sns.set_theme(style="whitegrid")
values = [20, 30, 50]
labels = ['Category A', 'Category B', 'Category C']

plt.pie(values, labels=labels, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Seaborn-Styled Pie Chart")
plt.show()
```

# 68(b). Scatter Plot

```python
sns.scatterplot(data=data, x='x', y='y', hue='category', palette='deep')
plt.title("Seaborn Scatter Plot")
plt.show()
```

# 68(c). Line Plot

```python
sns.lineplot(data=data, x='x', y='y', marker='o', color='blue')
plt.title("Seaborn Line Plot")
plt.show()
```

# 68(d). Histogram

```python
sns.histplot(data['y'], bins=10, kde=True, color='green')
plt.title("Seaborn Histogram with KDE")
plt.show()
```
